# Sales Prediction - Exploratory Data Analysis

Explore sales data and analyze factors influencing revenue.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

df = pd.read_csv('../data/sales.csv')
print(f"Dataset shape: {df.shape}")
df.head()
```

## Dataset Overview

```python
print("Dataset Info:")
print(df.info())
print("\nBasic Statistics:")
print(df.describe())

# Get sales column
sales_col = [col for col in df.columns if 'sales' in col.lower() or 'revenue' in col.lower()][0] if any('sales' in col.lower() or 'revenue' in col.lower() for col in df.columns) else df.columns[-1]
print(f"\nSales Statistics:")
print(f"Mean: ${df[sales_col].mean():,.2f}")
print(f"Median: ${df[sales_col].median():,.2f}")
print(f"Std Dev: ${df[sales_col].std():,.2f}")
```

## Sales Distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(df[sales_col], bins=30, color='#d62728', edgecolor='black')
axes[0].set_xlabel('Sales ($)')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Sales Distribution')

axes[1].boxplot(df[sales_col])
axes[1].set_ylabel('Sales ($)')
axes[1].set_title('Sales Box Plot')

plt.tight_layout()
plt.show()
```

## Feature Analysis

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Advertising vs Sales
if 'advertising' in df.columns:
    axes[0, 0].scatter(df['advertising'], df[sales_col], alpha=0.5, color='#ff7f0e')
    axes[0, 0].set_xlabel('Advertising ($K)')
    axes[0, 0].set_ylabel('Sales ($)')
    axes[0, 0].set_title('Advertising vs Sales')
    z = np.polyfit(df['advertising'], df[sales_col], 1)
    p = np.poly1d(z)
    axes[0, 0].plot(df['advertising'].sort_values(), p(df['advertising'].sort_values()), "r--", linewidth=2)

# Region analysis
if any('region' in col.lower() for col in df.columns):
    region_col = [col for col in df.columns if 'region' in col.lower()][0]
    region_sales = df.groupby(region_col)[sales_col].mean().sort_values(ascending=False)
    axes[0, 1].bar(range(len(region_sales)), region_sales.values, color='#1f77b4')
    axes[0, 1].set_xticks(range(len(region_sales)))
    axes[0, 1].set_xticklabels(region_sales.index, rotation=45)
    axes[0, 1].set_ylabel('Average Sales ($)')
    axes[0, 1].set_title('Average Sales by Region')

# Season analysis
if any('season' in col.lower() for col in df.columns):
    season_col = [col for col in df.columns if 'season' in col.lower()][0]
    season_sales = df.groupby(season_col)[sales_col].mean()
    axes[1, 0].bar(range(len(season_sales)), season_sales.values, color='#2ca02c')
    axes[1, 0].set_xticks(range(len(season_sales)))
    axes[1, 0].set_xticklabels(season_sales.index, rotation=45)
    axes[1, 0].set_ylabel('Average Sales ($)')
    axes[1, 0].set_title('Average Sales by Season')

axes[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

## Correlation Analysis

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
correlation = df[numeric_cols].corr()[sales_col].sort_values(ascending=False)

plt.figure(figsize=(10, 6))
correlation.plot(kind='barh', color='#1f77b4')
plt.xlabel('Correlation with Sales')
plt.title('Feature Correlation with Sales')
plt.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
plt.tight_layout()
plt.show()

print(correlation)
```

## Key Insights

1. **Advertising Impact**: Strong positive correlation between advertising spend and sales
2. **Regional Variations**: Some regions show significantly higher average sales
3. **Seasonal Patterns**: Clear seasonal trends in sales data
4. **Customer Base**: Customer count positively correlates with sales
5. **Non-linear Relationship**: Sales may not grow linearly with all inputs
